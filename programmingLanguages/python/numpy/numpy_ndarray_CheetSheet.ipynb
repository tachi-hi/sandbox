```python
b = [[[i * j * c for c in range(3)] for j in range(10)] for i in range(20)]
b = np.asarray(b,dtype=np.float32)
```

```python
img = plt.imshow(b/np.max(b))
```

```python
b.shape
```

```python
#swap
b_tmp = np.swapaxes(b, 0, 2)
b_tmp.shape
```

```python
# more useful form
b_tmp2 = np.einsum('ijk->kij', b)
b_tmp2.shape
```

```python
# charcters other than ijk are acceptable
b_tmp3 = np.einsum('abc->cab', b)
b_tmp3.shape
```
